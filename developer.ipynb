```python
from langchain_community.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings

# Beágyazó modell (Embedding) inicializálása
# Ez alakítja a mondatokat 384 dimenziós vektorokká
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
```

```python
from sentence_transformers import CrossEncoder
import numpy as np

# Cross-Encoder modell betöltése (Ez a Reranker-ünk)
reranker = CrossEncoder('BAAI/bge-reranker-v2-m3')
```

```python
# Metadata: az eredeti (változatlan) user_profile_inputs azonos indexű eleme
metadatas = [
                {"orig_word": user_profile_inputs[i]}   
                for i in range(len(user_profile_inputs_v2))
            ]

# Vektor adatbázis (Chroma) létrehozása a bővített szövegekből
vector_db = Chroma.from_texts(user_profile_inputs_v2, embeddings, metadatas=metadatas)
print("Adatbázis sikeresen létrehozva!")
```

```python
#user_input = "Elvált tőlem a párom. Mit tegyek, hogy könnyen ki tudjam ezt heverni?"
user_input = "Már 9 van és már vacsoráztam, de még mindig ennék valamit. Mit tanácsolsz?"
#user_input = "Elvált tőlem a párom. Mit tegyek, hogy könnyen ki tudjam ezt heverni?"
#user_input = "Elvált tőlem a párom. Mit tegyek, hogy könnyen ki tudjam ezt heverni?"
#user_input = "Elvált tőlem a párom. Mit tegyek, hogy könnyen ki tudjam ezt heverni?"
```

```python
# 1. Lekérünk TÖBB dokumentumot (Magas Recall)
alap_talalatok = vector_db.similarity_search(user_input, k=10)

# 2. Párokat képezünk a kérdésből és a megtalált dokumentumokból
parok = [[user_input, doc.page_content] for doc in alap_talalatok]

# 3. A Reranker pontozza a párokat (Mennyire válaszolja meg a dok. a kérdést?)
pontszamok = reranker.predict(parok)

# 4. Sorba rendezzük a dokumentumokat az új pontszámok alapján (csökkenő sorrendbe)
rendezett_indexek = np.argsort(pontszamok)[::-1]

rendezett_pontszamok = pontszamok[rendezett_indexek]

# 5. Új lista: csak az újrarendezett Top X találat
top10_reranked_talalatok = [alap_talalatok[idx] for idx in rendezett_indexek[:10]]
```

```python
print("--- SIMILARITY SEARCH TOP 10 (RERANK ELŐTT) ---")
for i, doc in enumerate(alap_talalatok[:10], start=1):
    print(f"{i}. {doc.metadata['orig_word']}")
```

```python
print("--- RERANK UTÁNI TOP 10 ---")
for i, idx in enumerate(rendezett_indexek[:10], start=1):
    doc = alap_talalatok[idx]
    print(f"{i}. pontszám: {pontszamok[idx]:.6f} | {doc.metadata['orig_word']}")
```

```python
print("--- HALADÓ RAG (RERANKED) TOP 5 TALÁLATOK ---")
for i, doc in enumerate(top10_reranked_talalatok, start=1):
    #print(doc.metadata["orig_word"])
    #original_idx = next(j for j, d in enumerate(alap_talalatok) if d.page_content == doc.page_content)
    print(f"{i}. találat (Pontszám: {rendezett_pontszamok[i]:.6f}): {doc.metadata["orig_word"]}")
```
